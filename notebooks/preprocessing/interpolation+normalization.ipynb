```python
"""
Notebook interpolation(linear) + normalization (ASWN) â€“ ampiimts Python 3.11 package

Goal:
Develop a universal method of pre processing on time 
series with missing values via the classical interpolation and regular 
time (linear method), normalization (ASWN method). 
"""
#file path without format
path = "../../dataset/ambient_temperature_system_failure"


%matplotlib widget
# ---- Package imports ----
import time
import pandas as pd
import numpy as np
import ampiimts
```

```python
df = pd.read_csv(path+".csv")
# df.head(-1)
```

```python
df_missing = ampiimts.missing_values(df, percent_missing=0.1)
# df_missing.head(-1)
```

```python
start_time = time.perf_counter()
df_normalize = ampiimts.pre_processed(df_missing, gap_multiplier=15, alpha=0.25)
end_time = time.perf_counter()
print(end_time - start_time)
# df_normalize.head(-1)
```

```python
ampiimts.plot_multiple_dfs([df, df_missing, df_normalize], ['Original', 'Missing', 'Normalize'])
```
